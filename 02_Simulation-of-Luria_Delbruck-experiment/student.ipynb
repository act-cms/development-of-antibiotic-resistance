Before we can get to perform a simulation of the actual experiment, let's see how we can model bacterial growth in a single generation.



Generate a simulation of a population of 100 bacteria, of which 90 are wild-type and 10 are mutant.

```python
# Generate initial population
population_size = 100
wt_percentage = 0.9
mutant_percentage = 1 - wt_percentage

num_wt = int(population_size * wt_percentage)
num_mutant = population_size - num_wt

initial_population = np.array([1] * num_wt + [0] * num_mutant)
np.random.shuffle(initial_population) # Shuffle the population

print("Initial Population:", initial_population)
print("Initial WT count:", np.sum(initial_population))
print("Initial Mutant count:", len(initial_population) - np.sum(initial_population))
```

Now double the population (as would happen if the cells divided).

```python
# Double the population
doubled_population = np.tile(initial_population, 2)

print("\nDoubled Population:", doubled_population)
print("Doubled WT count:", np.sum(doubled_population))
print("Doubled Mutant count:", len(doubled_population) - np.sum(doubled_population))
```

Let's switch one element randomly (e.g. a wild-type to mutant or a mutant back to wild-type)

```python
# Switch one element
index_to_switch = np.random.randint(0, len(doubled_population))
original_value = doubled_population[index_to_switch]

if original_value == 1:
    doubled_population[index_to_switch] = 0
    print(f"\nSwitched element at index {index_to_switch} from 1 to 0.")
else:
    doubled_population[index_to_switch] = 1
    print(f"\nSwitched element at index {index_to_switch} from 0 to 1.")

print("Population after switching one element:", doubled_population)
print("Final WT count:", np.sum(doubled_population))
print("Final Mutant count:", len(doubled_population) - np.sum(doubled_population))
```
