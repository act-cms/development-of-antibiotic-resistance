Another hypothesis that Luria and Delbruck proposed was that resistance mutations did not arise die to interactions with the phage. Instead, the mutant bacteria were already present **prior** to the interaction with the virus, and they became prominent only after the phage killed all the resistant cells. This is more in line with the Darwinian idea of evolution via natural selection (as opposed to the Damarckian concept of selection-dependent mutations): mutations happen independent of selective forces.
In this case, the number of mutant bacteria would be different based on **when** the mutation occurred.


*   If the mutants arose **early**, a single resistant cell would have divided many times before being plated on the agar containing the phages. In this scenario, one would observe a **large number** of resistant colonies on the plate.
*   If the mutant arose **late**, e.g. in the last division before the cells were plated on the lawn of phages, the number of resistant colonies would be much smaller.




This model is *continuous*. In this, susceptible bacteria grow at a rate *r*, but a small fraction of the offspring, *μ*, mutate and acquire resistance. The mutant bacteria grow just as well. Assuming that there is no link between resistance and growth rate, the mutants would also grow at a rate *r*. In this model then, we have to deal with two populations, *S* and *m*, the susceptible and mutant ones.

$$
\frac{dS}{dt}=rS(1-μ)
$$


$$
\frac{dm}{dt}=μrS+rm
$$

The first equation mathematically describes the growth rate of susceptible bacteria. In the second equation, the first term (*μrS*) is the number of new mutations arising and the second term (*rS*) is the growth of the mutants.

This is a dynamic system and quantifying both the susceptible and the mutant cells as a function of time with a given *μ* is challenging. Trying to solve this system of derivatives becomes simpler if you add them:

$$
\frac{dS}{dt} + \frac{dm}{dt} = r(S+m)
$$

This means that the entire population (let's call it *N*, which is equal to *S+m*) is growing at a rate *r*. So the solution to the exponential growth rate  for all cells is *N(t)=N<sub>0</sub>e<sup>rt</sup>*.

Returning to the original equations, and given that *m(t) = N(t) - S(t)*, we get:

$$
S(t) = N_0e^{r(1-μ)t}
$$

$$
m(t) = N_0e^{rt}(1-e^{-rμt})
$$



Given that an experiment like the one by Luria and Delbruck last about 10-20 generations, the value *rt* is close to that order of magnitude. The value *e<sup>-rμt</sup>* then can be approximated to *1 - rμt*, because *μ* << 1.
This means that the mutants will grow faster than exponentially.

$$
m(t) = N_0e^{rt}μrt
$$

Given the objective of the experiment, i.e. determining whether mutations arise spontaneously or are dependent on selection, the *r* number does not truly matter much; the product *rt* is more critical. Indeed, *r* is measured in inverse time, and *t* is time. If the units are the same, their product *rt* will have the same value. We can then define a new time scale *τ=rt*, and now d*τ*=*r*d*t*. Given this, τ=1 means 1 cell division, τ=20 means 20 cell divisions, etc. So, with this in mind, the above equations:

$$
\frac{dS}{dt}=rS(1-μ)
$$

$$
\frac{dm}{dt}=μrS+rm
$$

can now be divided by *r* on both sides to yield:

$$
\frac{dS}{rdt}=S(1-μ)
$$

$$
\frac{dm}{rdt}=μS+m
$$

Replacing d*τ*=*r*d*t*, will turn them to:

$$
\frac{dS}{dτ}=S(1-μ)
$$

$$
\frac{dm}{dτ}=μS+m
$$

This set of equations means that no matter what the growth rate, the sensitive population will grow more slowly than of the mutant population (the resistant bacteria).
Similar to what we did before, we can then get:

$$
m(τ)=N_0e^τμτ
$$

and at the end of the experiment, when τ=τ<sub>f</sub>, the number of mutants will be:

$$
m(τ_f)=N_fμτ_f
$$

Using the average number of mutants in replicate experiments, we can estimate the mutation rate:

$$
\hatμ=-\frac{\bar{m}_{obs}}{N_fτ_φ}.
$$

If mutants emerge prior to the selective pressure, as the spontaneous mutant hypothesis assumes, then the time of mutation appearence is important to the experimental outcome. For instance, if a single mutant cell develops five generations prior to the selection agent, then at the time of selection, there would be *2<sup>5</sup>=32* mutant cells which would be visible as colonies on the agar plates. If the mutant appeared 10 generations prior to the application of the selective pressure, the number of colonies would be *2<sup>10</sup>=1024*.
Interestingly, however, mutations are more likely to appear later in the culture, simply due to the sheer number of bacteria present. There are more bacteria later in the culture than earlier in the experiment.
The above dynamics, while insightful, set a trap. The continuous model implies that you can have fractional organisms grow. For instance, a value of *m=0.01* is not useful, as you will never have 0.01 mutant cells in an experiment. Lea and Coulson (1949) identified this issue and developed dynamics for a *discrete model*.
